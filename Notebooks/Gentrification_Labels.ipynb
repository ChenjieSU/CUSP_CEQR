```
#Census data for 2010
demo_data_2010 = pd.read_csv('acs_ct_2010.csv', usecols=['geo_id','total_pop','median_income','median_rent','owner_occupied_housing_units_median_value','bachelors_degree','white_pop','black_pop','hispanic_pop','asian_pop','amerindian_pop','other_race_pop','two_or_more_races_pop'])
demo_data_2010 = demo_data_2010.rename(columns = {'total_pop':'total_pop_10','median_income':'median_income_10','median_rent':'median_rent_10','owner_occupied_housing_units_median_value':'median_home_value_10','bachelors_degree':'bachelors_degree_10','white_pop':'white_pop_10','black_pop':'black_pop_10','hispanic_pop':'hispanic_pop_10','asian_pop':'asian_pop_10','amerindian_pop':'amerindian_pop_10','other_race_pop':'other_race_pop_10','two_or_more_races_pop':'two_or_more_races_pop_10'})
demo_data_2010.head()
```

```
#Census data for 2000
demo_data_2000 = pd.read_csv('census_2000.csv')
demo_data_2000 = demo_data_2000.drop(columns='Unnamed: 0')
demo_data_2000.head()
```

```
#Census data for 2013
demo_data_2013 = pd.read_csv('acs_ct_2013.csv', usecols=['geo_id','total_pop','median_income','median_rent','owner_occupied_housing_units_median_value','bachelors_degree','white_pop','black_pop','hispanic_pop','asian_pop','amerindian_pop','other_race_pop','two_or_more_races_pop'])
demo_data_2013 = demo_data_2013.rename(columns = {'total_pop':'total_pop_13','median_income':'median_income_13','median_rent':'median_rent_13','owner_occupied_housing_units_median_value':'median_home_value_13','bachelors_degree':'bachelors_degree_13','white_pop':'white_pop_13','black_pop':'black_pop_13','hispanic_pop':'hispanic_pop_13','asian_pop':'asian_pop_13','amerindian_pop':'amerindian_pop_13','other_race_pop':'other_race_pop_13','two_or_more_races_pop':'two_or_more_races_pop_13'})
demo_data_2013.head()
```

```
#Census data for 2018
demo_data_2018 = pd.read_csv('acs_ct_2018.csv', usecols=['geo_id','total_pop','median_income','median_rent','owner_occupied_housing_units_median_value','bachelors_degree','white_pop','black_pop','hispanic_pop','asian_pop','amerindian_pop','other_race_pop','two_or_more_races_pop'])
demo_data_2018 = demo_data_2018.rename(columns = {'total_pop':'total_pop_18','median_income':'median_income_18','median_rent':'median_rent_18','owner_occupied_housing_units_median_value':'median_home_value_18','bachelors_degree':'bachelors_degree_18','white_pop':'white_pop_18','black_pop':'black_pop_18','hispanic_pop':'hispanic_pop_18','asian_pop':'asian_pop_18','amerindian_pop':'amerindian_pop_18','other_race_pop':'other_race_pop_18','two_or_more_races_pop':'two_or_more_races_pop_18'})
demo_data_2018.head()
```

```
#Merging census data for all years together
demo_data_2000 = demo_data_2000.rename(columns={'GEOID':'geo_id'})
demo_data = pd.merge(demo_data_2018, demo_data_2013, on='geo_id',how='left')
demo_data = pd.merge(demo_data, demo_data_2010, on='geo_id',how='left')
demo_data = pd.merge(demo_data, demo_data_2000, on='geo_id',how='left')
demo_data.head()
```

```
#Subsetting census data for only NYC Census Tracts
demo_data = demo_data[demo_data['geo_id'].isin(ny_geoid)]
demo_data.shape
```

```
demo_data_10 = demo_data.copy()
demo_data_00 = demo_data.copy()
```

```
#Replacing NA values with 0
demo_data['median_home_value_13'] = demo_data['median_home_value_13'].replace(0, pd.np.nan)
```

## Difference between 2018 and 2013

```
#Calculating change in educational attainment
demo_data['educational_attainment_pct'] = (
    100 * demo_data['bachelors_degree_18'] /
    demo_data['total_pop_18']
)

demo_data['educational_attainment_pct13'] = (
    100 * demo_data['bachelors_degree_13'] / 
    demo_data['total_pop_13']
)

demo_data['educational_attainment_change'] = (
    demo_data['educational_attainment_pct'] - 
    demo_data['educational_attainment_pct13']
)
```

```
#Getting inflation rate for 2013
inflation_rate  = 1.08
```

```
#Getting median income and house value for the botton 40th percentile amongst all census tracts
median_home_2013_bottom_40th = demo_data['median_home_value_13'].quantile(0.4)

median_income_2013_bottom_40th = demo_data['median_income_13'].quantile(0.4)

#Calculating change in house prices
demo_data['home_pct_change'] = (
    (
        demo_data['median_home_value_18'] - # current median home value
        (demo_data['median_home_value_13'] * inflation_rate)  # previous median home value adjusted for inflation
    ) / 
    demo_data['median_home_value_13'] # current median home value 
)

median_home_change_top3rd = demo_data[demo_data['home_pct_change'] !=None ]['home_pct_change'].quantile(0.66)

print("Bottom 40th percentile household income threshold for metro level data: " , median_home_2013_bottom_40th)
print("Bottom 40th percentile median home value for metro level data: " , median_income_2013_bottom_40th)
print("Top third percentile for all-metro-level percentage increase in a tract’s inflation-adjusted median home value: " , median_home_change_top3rd)
```

```
#Considering top quantiles of educational attainment change
educational_attainment_threshold = demo_data['educational_attainment_change'].quantile(2.0/3)
print(demo_data['educational_attainment_change'].min())
print(demo_data['educational_attainment_change'].max())
print(educational_attainment_threshold)
```

```
# The tract had a population of at least 500 residents at the beginning and 
# end of a decade and was located within a central city
demo_data['has_pop_500+_2013'] = demo_data['total_pop_13'] >= 500
demo_data['has_pop_500+_2018'] = demo_data['total_pop_18'] >= 500

# The tract’s median household income was in the bottom 40th percentile when compared 
# to all tracts within its metro area at the beginning of the decade.
demo_data['median_40th_home_2013'] = demo_data['median_home_value_13'] < median_home_2013_bottom_40th

# The tract’s median home value was in the bottom 40th percentile when compared to all tracts within its metro area at the beginning of the decade.
demo_data['median_40th_income_2013'] = demo_data['median_income_13'] < median_income_2013_bottom_40th
demo_data.head(5)
```

```
# An increase in a tract's educational attainment, as measured by the 
# percentage of residents age 25 and over holding bachelor’s degrees, 
# was in the top third percentile of all tracts within a metro area.
demo_data['educational_is_upper3rd'] = demo_data['educational_attainment_change'] > educational_attainment_threshold

# The median home value increased, after accounting for inflation
demo_data['median_home_higher'] = demo_data['median_home_value_18'] > (demo_data['median_home_value_13'] * inflation_rate)

# The percentage increase in a tract’s inflation-adjusted median home 
# value was in the top third percentile of all tracts within a metro area.
demo_data['median_home_top_third'] = demo_data['home_pct_change'] > median_home_change_top3rd

demo_data.head()
```

```
#Creating logic for gentrification
demo_data['gentrified'] = (
    (demo_data['has_pop_500+_2013'] == True) & 
    (demo_data['has_pop_500+_2018'] == True) & 
    (demo_data['median_40th_home_2013'] == True) & 
    (demo_data['median_40th_income_2013'] == True) & 
    (demo_data['educational_is_upper3rd'] == True) & 
    (demo_data['median_home_higher'] == True) & 
    (demo_data['median_home_top_third'] == True)
)
```

```
#Calculating changes in femographics
demo_data['pct_white_alone_change'] = ((demo_data['white_pop_18']/demo_data['total_pop_18'])*100)-((demo_data['white_pop_13']/demo_data['total_pop_13'])*100)
demo_data['pct_white_alone_2018']= ((demo_data['white_pop_18']/demo_data['total_pop_18'])*100)

demo_data['pct_black_alone_change'] = ((demo_data['black_pop_18']/demo_data['total_pop_18'])*100)-((demo_data['black_pop_13']/demo_data['total_pop_13'])*100)
demo_data['pct_black_alone_2018'] = ((demo_data['black_pop_18']/demo_data['total_pop_18'])*100)

demo_data['pct_asian_alone_change'] = ((demo_data['asian_pop_18']/demo_data['total_pop_18'])*100)-((demo_data['asian_pop_13']/demo_data['total_pop_13'])*100)
demo_data['pct_asian_alone_2018'] = ((demo_data['asian_pop_18']/demo_data['total_pop_18'])*100)

demo_data['pct_hispanic_alone_change'] = ((demo_data['hispanic_pop_18']/demo_data['total_pop_18'])*100)-((demo_data['hispanic_pop_13']/demo_data['total_pop_13'])*100)
demo_data['pct_hispanic_alone_2018'] = ((demo_data['hispanic_pop_18']/demo_data['total_pop_18'])*100)
```

```
demo_data["low_population"] = (demo_data["total_pop_13"] < 500) | (demo_data["total_pop_18"] < 500)
```

```
#Calculating if census tract is eligible for gentrification in 2013
demo_data["eligible_for_gentrification"] = (
  (demo_data["low_population"] == False) &
  (demo_data["median_home_value_13"] < median_home_2013_bottom_40th) &
  (demo_data["median_income_13"] < median_income_2013_bottom_40th)
)
```

```
#Calculating if census tract is not eligible for gentrification in 2013
demo_data["not_eligible_highpop"] = (
  (demo_data["low_population"] == False) &
  (demo_data["eligible_for_gentrification"] == False) 
)

len(demo_data[demo_data["not_eligible_highpop"]])
```

```
demo_data['eligible_for_gentrification'].value_counts()
```

```
demo_data.to_csv('gentrification_13_18.csv')
```
